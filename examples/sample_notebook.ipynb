# My very important analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.DataFrame({'x': np.random.normal(size=200),
                     'y': np.random.normal(size=200)})

fig, ax = plt.subplots(figsize=(5, 5))
data.plot(x='x', y='y', kind='scatter', ax=ax, alpha=.5, s=300)
ax.set_title("My very important graph")
```
